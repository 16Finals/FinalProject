```python
import re
import pandas as pd

# ────────────────────────────────────────────
# 1) 서열(순서) 컬럼 목록
# ────────────────────────────────────────────
ord_cols = [
    '인입횟수_ARS_R6M',
    '이용메뉴건수_ARS_R6M',
    '방문횟수_PC_R6M',
    '방문일수_PC_R6M',
    '방문횟수_앱_R6M'
]

# ────────────────────────────────────────────
# 2) 인코딩 함수  ―  df[col]을 숫자로 **덮어쓰기**
# ────────────────────────────────────────────
def ordinal_encode_inplace(df: pd.DataFrame, col: str):
    """
    '1회 이상', '10회 이상' → 0,1,2… 로 변환
    숫자가 없는 값은 -1
    """
    # (1) 고유 카테고리 추출
    uniques = df8[col].dropna().unique()

    # (2) '숫자' 추출  → 정렬 키
    def extract_num(s):
        m = re.search(r'\d+', str(s))
        return int(m.group()) if m else float('inf')   # 숫자 없으면 inf
    
    ordered = sorted(uniques, key=extract_num)

    # (3) 매핑 딕셔너리 생성
    mapping = {cat: idx for idx, cat in enumerate(ordered) if extract_num(cat) != float('inf')}
    mapping.update({cat: -1 for cat in ordered if extract_num(cat) == float('inf')})

    # (4) 실제 덮어쓰기
    df8[col] = df8[col].map(mapping).astype('int8')

    print(f"[{col}] 매핑 ▶ {mapping}")

# ────────────────────────────────────────────
# 3) 모든 서열 컬럼 인코딩
# ────────────────────────────────────────────
for c in ord_cols:
    ordinal_encode_inplace(df8, c)

# ────────────────────────────────────────────
# 4) 확인
# ────────────────────────────────────────────
print(df8[ord_cols].head())
print(df8[ord_cols].dtypes)   # int8 로 변경된 것 확인
```
