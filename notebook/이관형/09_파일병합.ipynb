```python
import pandas as pd
import gc
from google.colab import drive
drive.mount('/content/drive')

data_root = '/content/drive/MyDrive/부트캠프/'
data_path = data_root + '11_파이널/'
```

```python
df1 = pd.read_parquet(data_path + 'data/train1.parquet')
df1.info()
```

```python
df1[df1.select_dtypes(include='int64').columns] = df1.select_dtypes(include='int64').astype('int32')
df1.info()
```

```python
gc.collect()
df4 = pd.read_parquet(data_path + 'data/train4.parquet')
df4.info()
```

```python
dd.loc['max'].sort_values()
```
