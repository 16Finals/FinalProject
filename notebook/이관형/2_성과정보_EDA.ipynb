```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def calculate_vif(X):
    X = add_constant(X)  # 상수항 추가 (intercept)
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
numeric_cols = df88.select_dtypes(include='number')
hrm = df88['혜택수혜율_R3M'].median()
hbm = df88['혜택수혜율_B0M'].median()
numeric_cols['혜택수혜율_R3M'].fillna(hrm, inplace=True)
numeric_cols['혜택수혜율_B0M'].fillna(hbm, inplace=True)
vif_result = calculate_vif(numeric_cols)
print(vif_result)
```

```python
rests = df88.iloc[:, 20:28]
rests.columns
```

```python
rests.corr()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

ssc = StandardScaler()
scaled_rest = ssc.fit_transform(rests)

pca = PCA(n_components=1)
pca.fit(scaled_rest)
rest_pca = pca.transform(scaled_rest)

# 주성분 분석 결과 출력
print("주성분 비율:", pca.explained_variance_ratio_)
print("주성분 벡터 (components):\n", pca.components_)

# 주성분을 기준으로 변환된 데이터
transformed_data = pca.transform(scaled_rest)

# 변환된 데이터 출력
print("변환된 데이터 (주성분 공간):\n", transformed_data)
```
