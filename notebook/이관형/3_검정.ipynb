```python
import pandas as pd
import gc
from google.colab import drive
drive.mount('/content/drive')

data_root = '/content/drive/MyDrive/부트캠프/'
data_path = data_root + '11_파이널/'
```

```python
df8 = pd.read_parquet(data_path + 'processed/df8.parquet')
df8
```

```python
cardloan = df8[['증감율_이용건수_카드론_전월','증감율_이용금액_카드론_전월']]
cardloan.head()
```

```python
dm = pd.get_dummies(df8['Segment'])
dm
```

```python
dm['증감건수0'] = cardloan['증감율_이용건수_카드론_전월'] == 0
dm['증감금액0'] = cardloan['증감율_이용금액_카드론_전월'] == 0
dm
```

```python
ct = pd.crosstab(dm['E'], dm['증감금액0'])
ct
```

```python
import scipy.stats as stats

# 피셔의 정확한 검정 수행
oddsratio, p_value = stats.fisher_exact(ct)

# p-value 출력
print(f"P-value: {p_value}")
```

```python
ct = pd.crosstab(dm['A'], dm['증감금액0'])

# 피셔의 정확한 검정 수행
oddsratio, p_value = stats.fisher_exact(ct)

# p-value 출력
print(f"P-value: {p_value}")
```
