```python
segment_abc = segment[segment['abc'] == 1]
y = segment_abc['Segment']
del segment_abc
del segment
gc.collect()
y
```

```python
train_abc.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
```

```python
x = train_abc.drop(['ID','기준년월'], axis=1)
y_one = y == 'C'
y_one = y_one.astype(int)
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train1, x_val, y_train1, y_val = train_test_split(x_scaled, y_one, test_size=0.2, stratify=y_one, random_state=42)
x_train, x_opt, y_train, y_opt = train_test_split(x_train1, y_train1, test_size=0.2, stratify=y_train1, random_state=42)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import f1_score

# Objective 함수 정의
def xgb_objective(trial):
    # 탐색할 하이퍼파라미터 정의
    params = {
        "booster": "dart",
        "max_depth": 4,
        "max_leaves": 96,
        "early_stopping_rounds": 20,
        "n_estimators": 1000,
        "device": "cuda",
        "tree_method": "hist",
        "silent": 1,
    }

    clf = XGBClassifier(**params)
    clf.fit(
        x_train,
        y_train,
        eval_set=[(x_opt, y_opt)],
    )

    preds = clf.predict(x_val)
    f1 = f1_score(y_val, preds, average='macro')
    return f1
```

```python
import optuna
# Optuna 실행
study = optuna.create_study(direction="maximize")
study.optimize(xgb_objective, n_trials=10)

print("Best trial:")
trial = study.best_trial
print(f"  F1 Score: {trial.value}")
print("  Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")
```

```python
model = XGBClassifier(**trial.params, random_state=42)
model.fit(x_train, y_train)
```

```python
y_val_pred = model.predict(x_val)
f1_micro = f1_score(y_val, y_val_pred, average='macro')
print(f"F1 score (macro): {f1_micro:.4f}")
```
