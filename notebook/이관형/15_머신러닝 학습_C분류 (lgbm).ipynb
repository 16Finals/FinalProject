```python
train_abc.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
```

```python
x = train_abc.drop(['ID','기준년월'], axis=1)
y_one = y == 'C'
y_one = y_one.astype(int)
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train1, x_val, y_train1, y_val = train_test_split(x_scaled, y_one, test_size=0.2, random_state=42)
x_train, x_opt, y_train, y_opt = train_test_split(x_train1, y_train1, test_size=0.2, random_state=42)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import f1_score
import optuna

# Objective 함수 정의
def lgbm_objective(trial):
    # 탐색할 하이퍼파라미터 정의
    params = {
        "booster": "dart",
        "max_depth": 12,
        "max_leaves": 180,
        "row_sample_rate": 0.6,
        "colsample_bytree": 0.9,
        "early_stopping_rounds": 50,
        "n_estimators": 1000,
    }

    clf = LGBMClassifier(**params)
    clf.fit(
        x_train,
        y_train,
        eval_set=[(x_opt, y_opt)],
    )

    preds = clf.predict(x_val)
    f1 = f1_score(y_val, preds, average='macro')
    return f1
```

```python
# Optuna 실행
study = optuna.create_study(direction="maximize")
study.optimize(lgbm_objective, n_trials=10)

print("Best trial:")
trial = study.best_trial
print(f"  F1 Score: {trial.value}")
print("  Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")
```
